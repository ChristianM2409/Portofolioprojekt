```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

from core.data import load_from_kaggle
```

```python
dataset_link = "samruddhi4040/hr-data-analytics" # replace with your dataset link from Kaggle 
destination = "../data/raw"
dataset_name = dataset_link.split("/")[-1]

files = load_from_kaggle(
    dataset_link=dataset_link, 
    destination=destination,
    )
```

```python
files
```

```python
df = pd.read_csv("/".join(["../data/raw/", dataset_name, files[0]]))
df.head()
```

```python
display(
    "Shape",
    df.shape,
    "Description",
    df.describe().round(2),
    "Duplicates",
    df.duplicated().sum(),
)


pd.DataFrame(
    {
        "Data Types": df.dtypes,
        "Missing Values": df.isnull().sum(),
        "Unique Values": df.nunique(),
        "Sample Values": [df[col].sample(3).tolist() for col in df.columns]
    })
```

```python
fig, ax = plt.subplots(figsize=(12, 6
                                ), dpi=200)

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", center=0, ax=ax);
```

```python
#change object in categories
for cat_name in ["EmpID", "AgeGroup"]: #, "Attrition","Department", "EducationField","Gender", "JobRole","MaritalStatus","SalarySlab", "Over18","OverTime"):
    df.loc[:,cat_name] = df.loc[:,cat_name].astype("category")
print(df.dtypes)
```

```python
# Change object in category
for col in ["EmpID", "AgeGroup","BusinessTravel", "Attrition","Department", "EducationField","Gender", "JobRole","MaritalStatus","SalarySlab", "Over18","OverTime", "JobSatisfactionText"]:
    df[col] = df[col].astype("category")

df.dtypes
```

```python
#number of duplicate rows
df.duplicated().sum()
```

```python
#show duplicated rows
df[df.duplicated(keep=False)]
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
#check missing value
df_NaN_True = df.isna()
mask_NaN=df_NaN_True.sum()

col_NaN=mask_NaN[mask_NaN !=0]

print(col_NaN)
```

```python
df[df.isnull().any(axis=1)]
```

```python
median_value_YearsWithCurrManager=df["YearsWithCurrManager"].median()
df["YearsWithCurrManager"]=df["YearsWithCurrManager"].fillna(median_value_YearsWithCurrManager)

```

```python
mapping = {
    1: "Very Dissatisfied",
    2: "Dissatisfied",
    3: "Satisfied",
    4: "Very Satisfied"
}

df["JobSatisfactionText"] = df["JobSatisfaction"].map(mapping)
```

```python
#EDA Check alle category spalten
# Alle Spalten, die 'object' oder 'category' sind
cat_columns = df.select_dtypes(include=['category']).columns
print(cat_columns)

int_columns = df.select_dtypes(include=['int']).columns
print(int_columns)
```

```python
#EDA Check AgeGroup
# Kreuztabelle: Anteil der Fluktuation pro Altersgruppe
pd_AgeGroup_Attrition_relation = pd.crosstab(
    index=df["AgeGroup"],
    columns=df["Attrition"],
    normalize="index"
)
print(pd_AgeGroup_Attrition_relation)

# Only "Yes"
pd_AgeGroup_Attrition_relation_yes = pd_AgeGroup_Attrition_relation[["Yes"]]



print(pd_AgeGroup_Attrition_relation_yes)

#Visualizing: chart focus on attrition "Yes" 
ax = pd_AgeGroup_Attrition_relation_yes.plot(
    kind="bar",
    title="Attrition by Age Group",
    legend=False
    #stacked=True,       # besser lesbar bei Anteilen
    #color=["#66b3ff"]  # optional: Farbanpassung
)

ax.set_xlabel("Age Group")
ax.set_ylabel("Attrition Rate")
ax.xaxis.set_tick_params(labelrotation=0)
```

```python
#EDA Check EducationField
pd_EducationField_Attrition_relation = pd.crosstab(index = df.loc[:,"EducationField"], 
                                    columns =df.loc[:,"Attrition"],normalize = "index" )
print(pd_EducationField_Attrition_relation)

#chart focus on attrition "Yes"

pd_EducationField_Attrition_relation_yes=pd_EducationField_Attrition_relation[["Yes"]]



ax=pd_EducationField_Attrition_relation_yes.plot(
    kind="bar",
    title= "Attrition by EducationField",
    figsize=(10,10),
    legend=False
    )
 
ax.set_ylabel("Attrition Rate")
ax.xaxis.set_tick_params(labelrotation=0)
```

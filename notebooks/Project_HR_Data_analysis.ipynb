```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

from core.data import load_from_kaggle
```

```python
dataset_link = "samruddhi4040/hr-data-analytics" # replace with your dataset link from Kaggle 
destination = "../data/raw"
dataset_name = dataset_link.split("/")[-1]

files = load_from_kaggle(
    dataset_link=dataset_link, 
    destination=destination,
    )
```

```python
files
```

```python
df = pd.read_csv("/".join(["../data/raw/", dataset_name, files[0]]))
df.head()
```

```python
display(
    "Shape",
    df.shape,
    "Description",
    df.describe().round(2),
    "Duplicates",
    df.duplicated().sum(),
)


pd.DataFrame(
    {
        "Data Types": df.dtypes,
        "Missing Values": df.isnull().sum(),
        "Unique Values": df.nunique(),
        "Sample Values": [df[col].sample(3).tolist() for col in df.columns]
    })
```

```python
fig, ax = plt.subplots(figsize=(12, 6
                                ), dpi=200)

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", center=0, ax=ax);
```

```python
# CAlculate the distribution
JobSatisfaction_tab = pd.crosstab(index=df["JobSatisfaction"], columns="count")



plt.figure(figsize=(6, 6))
plt.pie(
    JobSatisfaction_tab["count"],
    labels=JobSatisfaction_tab.index.astype(str),
    autopct="%1.1f%%"
)
plt.title("JobSatisfaction Distribution")
plt.show()
```

```python
# CAlculate the distribution
JobSatisfaction_tab = pd.crosstab(index=df["JobSatisfaction"], columns="count")

# Label is mapping
labels= JobSatisfaction_tab.index.map(mapping)

plt.figure(figsize=(6, 6))
plt.pie(
    JobSatisfaction_tab["count"],
    labels=labels,
    autopct="%1.1f%%"
)
plt.title("JobSatisfaction Distribution")
plt.show()
```

```python
# Count frequency of each JobSatisfaction value and sort by index (1–4)  
counts = df["JobSatisfaction"].value_counts().sort_index()  # ensures correct order

# Convert numeric values (1–4) to text labels using the mapping
labels = counts.index.map(mapping)  # replace numbers with descriptive labels

# Set custom colors: 1=red, 2=orange, 3=blue, 4=green
colors = ["red", "orange", "blue", "green"]  # color mapping in correct order

# Create pie chart
plt.figure(figsize=(6, 6))  # set figure size
plt.pie(
    counts,                   # values to plot
    labels=labels,            # mapped text labels
    colors=colors,            # custom color selection
    autopct="%1.1f%%"         # show percentage with 1 decimal
)
plt.title("Job Satisfaction Distribution")  # chart title
plt.show()  # display chart
```

```python
from numpy import unstack


df.groupby("Department")["Attrition"].value_counts(normalize=True).unstack()*100

attr = df.groupby("Department")["Attrition"] \
         .value_counts(normalize=True) \
         .unstack() * 100

plt.figure()
attr.plot(kind="bar")
plt.ylabel("Attrition Rate (%)")
plt.title("Attrition Rate by Department")
plt.xticks(rotation=45)
plt.legend(title="Attrition")
plt.tight_layout()
plt.show()

# absolute Werte zählen
attr_abs = df.groupby("Department")["Attrition"] \
             .value_counts() \
             .unstack(fill_value=0)

plt.figure()
attr_abs.plot(kind="bar")
plt.ylabel("Count")
plt.title("Attrition by Department (Absolute Numbers)")
plt.xticks(rotation=45)
plt.legend(title="Attrition")
plt.tight_layout()
plt.show()

```

```python
#EDA Check AgeGroup
pd_AgeGroup_Attrition_relation = pd.crosstab(index = df.loc[:,"AgeGroup"], 
                                    columns =df.loc[:,"Attrition"],normalize = "index" )
print(pd_AgeGroup_Attrition_relation)

import matplotlib.pyplot as plt


ax=pd_AgeGroup_Attrition_relation.plot(kind="bar",title="Attrition by AgeGroup") 
ax.xaxis.set_tick_params(labelrotation=0)
```

```python
#EDA Check EducationField
d_AgeGroup_Attrition_relation = pd.crosstab(
    index=df["AgeGroup"],
    columns=df["Attrition"],
    normalize="index"   # ergibt relative Anteile (Prozente)
)

print(pd_AgeGroup_Attrition_relation)

# Visualisierung als Balkendiagramm
ax = pd_AgeGroup_Attrition_relation.plot(
    kind="bar",
    figsize=(8, 5),
    title="Attrition by Age Group",
    stacked=True,       # besser lesbar bei Anteilen
    color=["#66b3ff", "#ff9999"]  # optional: Farbanpassung
)

ax.set_xlabel("Age Group")
ax.set_ylabel("Proportion")
ax.xaxis.set_tick_params(labelrotation=0)
plt.legend(title="Attrition")
plt.tight_layout()
plt.show()
```
